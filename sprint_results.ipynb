```python
# import libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'C:\Users\MDELANEY1\OneDrive - Anglo American\Documents\Personal\Data projects\formula 1 championship\raw_data\sprint_results.csv')
```

```python
df.head()
```

```python
# drop columns

df = df.drop(columns='positionText')
```

```python
df.info()
```

```python
# change datatypes

df['fastestLap'] = pd.to_numeric(df['fastestLap'], errors='coerce').fillna(0).astype('int64')

# custom function to convert time
def convert_time(t):
    # check if the value is not a number (NaN) or "\N"
    if pd.isna(t) or t == "\\N":
        return pd.NaT
    try:
        # try to convert using "minutes:seconds.milliseconds" format
        return pd.to_datetime(t, format='%M:%S.%f').time()
    except ValueError:
        # if error, try to convert using "hours:minutes:seconds.milliseconds" format
        return pd.to_datetime(t, format='%H:%M:%S.%f').time()

# apply function
df['fastestLapTime'] = df['fastestLapTime'].apply(convert_time)

# convert the 'milliseconds' column to numeric values, coercing errors to NaN
df['milliseconds'] = pd.to_numeric(df['milliseconds'], errors='coerce')

# convert to timedelta
df['milliseconds'] = pd.to_timedelta(df['milliseconds'], unit='ms')
```
