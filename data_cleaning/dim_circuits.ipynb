```python
# import libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# read csv into dataframe

df = pd.read_csv(r'circuits.csv')
df.head()
```

```python
# drop columns

df = df.drop(columns=['url'])
df.head()
```

```python
# check datatypes

df.info()
```

```python
# check for duplicates

duplicates = df.duplicated()

# print duplicate rows

print('Duplicate rows:')
print(df[duplicates])
```

```python
df['lat'] = df['lat'].astype(str).str.replace('.', ',')
df['lng'] = df['lng'].astype(str).str.replace('.', ',')

df.head()
```

```python
# save dataframe to csv

df.to_csv(r'dim_circuits.csv', index=False)
```
