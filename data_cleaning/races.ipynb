```python
# import libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r'races.csv')
```

```python
df.head()
```

```python
df = df.drop(columns='url')
```

```python
# custom function to convert date
def convert_date(d):
    if pd.isna(d) or d == "\\N":
        return None
    else:
        return pd.to_datetime(d)

# apply the function to each date column
for col in ['date', 'fp1_date', 'fp2_date', 'fp3_date', 'quali_date', 'sprint_date']:
    df[col] = df[col].apply(convert_date)
```

```python
df.info()
```

```python
# custom function to convert time
def convert_time(t):
    # check if the value is not a number (NaN) or "\N"
    if pd.isna(t) or t == "\\N":
        return pd.NaT
    try:
        # try to convert using "hours:minutes:seconds.milliseconds" format
        return pd.to_datetime(t, format='%H:%M:%S.%f').time()
    except ValueError:
        # if error, try to convert using "hours:minutes:seconds" format
        return pd.to_datetime(t, format='%H:%M:%S').time()
```
